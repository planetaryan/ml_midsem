```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig=plt.figure()

ax1=fig.add_axes([0,0,1,1])
ax2=fig.add_axes([0.2,0.5,0.2,0.2])

x=np.arange(10)
y=np.arange(10)

ax1.plot(x,y,lw=2,color='blue')
ax2.plot(x,y,lw=2,color='green')

plt.show()

```
