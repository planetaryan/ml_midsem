```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df=pd.read_csv(r"C:\Users\plane\Home\Desktop\ML_Project\week4\gold_price.csv")
df.head()
```

```python
n=len(df)
x_sum=np.sum(df['year'])
x_sq_sum=np.sum(df['year']**2)

y_sum=np.sum(df['price'])
xy_sum=np.sum(df['price']*df['year'])
```

```python
matA=np.array(([n,x_sum],
               [x_sum,x_sq_sum]))

matB=np.array(([y_sum],
               [xy_sum]))

coeffs=np.linalg.inv(matA)@matB

coeffs
```

```python
b0,b1=coeffs
```

```python
df['slr_pred']=b0+(b1*df['year'])
df[['slr_pred','price']]
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()

X=df[['year']]
y=df['price']

model.fit(X,y)

print(model.intercept_)
print(model.coef_)
```

```python
df=pd.read_csv(r"C:\Users\plane\Home\Desktop\ML_Project\week4\mlr_dataset\multiple_linear_regression_dataset.csv")
df.head()
```

```python
len(df)
```

```python
n=len(df)
x1_sum=np.sum(df['age'])
x2_sum=np.sum(df['experience'])
x1_2_sum=np.sum(df['age']**2)
x1x2_sum=np.sum(df['age']*df['experience'])
x2_2_sum=np.sum(df['experience']**2)

y_sum=np.sum(df['income'])
x1y_sum=np.sum(df['age']*df['income'])
x2y_sum=np.sum(df['experience']*df['income'])

matA=np.array(([n,x1_sum,x2_sum],
               [x1_sum,x1_2_sum,x1x2_sum],
               [x2_sum,x1x2_sum,x2_2_sum]))

matB=np.array(([y_sum],
               [x1y_sum],
               [x2y_sum]))

coeffs=np.linalg.inv(matA)@matB
coeffs
```

```python
b0,b1,b2=coeffs

df['mlr_pred']=b0+(b1*df['age'])+(b2*df['experience'])
df[['mlr_pred','income']]
```

```python
X=df[['age','experience']]
y=df['income']

model_mlr=LinearRegression()

model_mlr.fit(X,y)

print(model_mlr.intercept_)
print(model_mlr.coef_)
```
